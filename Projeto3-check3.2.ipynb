```python
#criando categorias
dados['waterfront']=dados['waterfront'].astype('category')
dados['waterfront'].cat.categories = (['Não','Sim'])
dados.head()

dados['condition']= dados['condition'].astype('category')
dados['condition'].cat.categories = ('Ruim','Satisfatório', 'Ok', 'Bom', 'Excelente')

def grading(dados):
    x=dados['grade']
    if x<=2: return 1
    elif x<=4: return 2
    elif x<=6: return 3
    elif x==7: return 4
    elif x<=9: return 5
    elif x<=11: return 6
    elif x<=13: return 7
    return 8

grade=dados.apply(grading,axis=1)
dados['grade']=grade
dados.grade=dados.grade.astype('category')
dados.grade.cat.categories=(['Péssimo','Muito Ruim','Ruim','Ok','Bom','Muito Bom', 'Excelente'])
```

Os dois superiores e inferiores a ele seguem uma sequência lógica de nota e classificação, indo de "Excelênte" a "Péssimo".

o MODELO DE REGRESSÃO LINEAR (Multiple Linear Regression)

        Para cada modelo a ser realizado, devem ser calculados o R2 e RMS, que são valores que ajudam a determinar a performance de cada modelo aplicado.
        O R2 é o coeficiente de determinação e o RMS, que é o quadrado da correlação entre o valor real e o valor predito. Ele varia entre 0 e 1, com isso, quanto mais próximo de 1, melhor, pois mostra uma maior correlação entre os valores. 
        O RMS é a raíz do erro médio quadrático. Logo, quanto menor o seu valor, mais próximo de 0, melhor, pois indica que a previsão corresponde ao valor real, visto que quase não hpa erro.

```python
#criando as funcoes para o calculo de R2  RMS

def calculor2 (y, f, y_):
    somac = ((y- f)**2).sum()
    somab = ((y - y_)**2).sum()
    r2 = 1- (somac/somab)
    return r2

def RMS (n, y, f):
    soma = ((y-f)**2).sum()
    rms = (soma/n)**(1/2)
    return rms
```

MODELO DE REGRESSÃO LINEAR (Multiple Linear Regression)

Um modelo simples de mínimos quadrados ordinários.

endog (array-like) - A variável dependente.

exog (array-like) - Um array nobs x k onde nobs é o número de observações e k é o número de regressores. Uma interceptação não é incluída por padrão e deve ser adicionada pelo usuário. Consulte statsmodels.tools.add_constant.

missing (str) - As opções disponíveis são "none", "drop" e "raise". Se "none", nenhuma verificação de nan é feita. Se "drop", quaisquer observações com nans são descartadas. Se "raise", um erro é levantado. O padrão é "none".

hasconst (None ou bool) - Indica se o RHS inclui uma constante fornecida pelo usuário. Se True, uma constante não é verificada e k_constant é definida como 1 e todas as estatísticas de resultado são calculadas como se uma constante estivesse presente. Se False, uma constante não é verificada e k_constant é definido como 0.


weights
escalar - Tem um atributo weights = array (1.0) devido à herança do WLS.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

EU TÔ USANDO A VARIÁVEL sqrt_lot  COMO TESTE.. TEMOS QUE VER CERTINHO QUAIS VARIÁVEIS TEREMOS QUE USAR PARA PLOTAR OS GRÁFICOS, SE OPTARMOS POR FAZER ESSA ANÁLISE PARCIAL E DEPOIS A MÚLTIPLA

```python
modelo_linear_com_uma_variável= ols("price ~ sqft_lot", data=casa).fit()
```

- Adj. R-squared indica que 65% do preço da casa pode ser explicado pelas variáveis escolhidas.

- O coeficiente de regressão (coeficiente de regressão) representa a mudança na variável dependente resultante de uma mudança de uma unidade na variável preditora, sendo que todas as outras variáveis, estão sendo mantidas constantes. 


- O erro padrão mede a precisão do coeficiente total_unumployed estimando a variação do coeficiente se o mesmo teste fosse executado em uma amostra diferente. 

- Um valor p baixo indica que os resultados são estatisticamente significativos, isto é, em geral, o valor p é menor que 0,05

- O intervalo de confiança é um intervalo dentro do qual o nosso coeficiente provavelmente cairá. 

Regressão PLOTS

O gráfico “Y e Ajustado vs. X” representa a variável dependente em relação aos nossos valores previstos com um intervalo de confiança

A relação inversa em nosso gráfico indica que variáveis estão negativamente correlacionados, ou seja, quando uma variável aumenta, a outra diminui.

O gráfico “Residuals versus uma variável” mostra os erros do nosso modelo versus a variável preditor especificada. Cada ponto é um valor observado; a linha representa a média desses valores observados. Como não há um padrão na distância entre os pontos e o valor médio, a suposição de homocedasticidade do OLS é válida.

O “Gráfico de regressão parcial” mostra a relação entre price e variável, levando em conta o impacto da adição de outras variáveis independentes em nosso coeficiente total_unemployed existente. Veremos mais adiante como esse mesmo gráfico muda quando adicionamos mais variáveis.

O gráfico Componente e Componente Plus Residual (CCPR) é uma extensão do gráfico de regressão parcial, mas mostra onde a nossa linha de tendência ficaria depois de adicionar o impacto de adicionar outras variáveis independentes ao nosso coeficiente da variável escolhida existente.

- Isso produz nossos quatro gráficos de regressão para cada variável


```python
# passar no modelo como o primeiro parâmetro, depois especifique a variável preditora que queremos analisar
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog (modelo_linear_com_uma_variável, "sqft_lot", fig = fig)
```


O próximo gráfico representa nossa linha de tendência (verde), as observações (pontos) e nosso intervalo de confiança (vermelho)

```python
# predictor variable (x) and dependent variable (y)
x = casa[['sqft_lot']]
y = casa[['price']]

# Retrieve our confidence interval values
# _ is a dummy variable since we don't actually use it for plotting but need it as a placeholder
# since wls_prediction_std(housing_model) returns 3 values
_, confidence_interval_lower, confidence_interval_upper = wls_prediction_std(modelo_linear_com_uma_variável)

fig, ax = plt.subplots(figsize=(10,7))

# plot the dots
# 'o' specifies the shape (circle), we can also use 'd' (diamonds), 's' (squares)
ax.plot(x, y, 'o', label="casa")

# plot the trend line
# g-- and r-- specify the color to use
ax.plot(x, modelo_linear_com_uma_variável.fittedvalues, 'g--.', label="OLS")

# plot upper and lower ci values
ax.plot(x, confidence_interval_upper, 'r--')
ax.plot(x, confidence_interval_lower, 'r--')

# plot legend
ax.legend(loc='best');
```
